Data Cleaning

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```python
# checking the outliers in independent variables

col = ['R', 'AB', 'H', '2B', '3B', 'HR', 'BB', 'SO', 'SB', 'RA', 'ER', 'ERA', 'CG', 'SHO', 'SV', 'E',]

for i in range(0,16):
    print(sns.boxplot(df[col[i]]))
    plt.show()
```

```python
# there are outliers in some variable(R, ERA, SHO, SV, E) so we have to remove them from the dataframe

from scipy.stats import zscore

z = np.abs(zscore(df))
df1 = df[(z<3).all(axis=1)]
print(df1.shape)
print(df.shape)

# One row is removed from the original dataset
```

Analysing the relationship between the Data

```python
# Using the regression plot- to analyse the linear regression model between all the independent and dependent variable
# y= mx+c is the linear regression model where m is the slope of the line
# we can remove the columns from the dataset which gives the slope value tends to zero because it will not impact the dataset

for i in range(0,16):
    print(sns.regplot(x=col[i], y='W', data=df1))
    plt.show()
```

```python
# from the regplot(), we can see that in some plot the slope value is tends to zero
# AB, H, CG, and E shows the slope value tends to zero
```

```python
# the regplot() graph already shows that some slopes are tends to zero like AB, H, CG, and E, now removing this.
# now droppping the columns which are showing correlation less than 0.1 from the dataset, considering row 1

np.abs(df1.corr())<0.1
```

```python
# from the 1st row we can see that AB, H, CG and E showing True so we are removing these row from the original dataset
#considering row 1

df2=df1.drop(['AB','H','CG','E'],axis=1)
df2.head()
```

```python
# W is our dependent variables and rest are the independent variables
# W = m1(R) + m2(2B) + m3(3B) +m4(HR) + m5(BB) + m6(SO) + ...
# W = Y_axis, rest are X_axis

df2_x=df2.iloc[:,1:13]
print(df2_x.head())

print('\n')

df2_y=df2['W']
print(df2_y.head())
```

```python
# checking the skewness in x variables

df2_x.skew()
```

```python
# SV is positively skewed

df2_x.skew()>0.55
```

```python
# Eliminating the skewed data

for i in df2_x.skew().index:
    if df2_x.skew().loc[i]>0.55:
        df2_x[i]=np.log1p(df2_x[i])

df2_x.skew()
```

```python
# now using standard scaler

from sklearn.preprocessing import StandardScaler

ss=StandardScaler()
ss.fit(df2_x)
x=ss.transform(df2_x)

x=pd.DataFrame(x,columns=df2_x.columns)
x.head()
```

```python
# R2 score value to fit the model

from sklearn import linear_model
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
max_r_score=0
for r_state in range(50,100):
    x_train, x_test, y_train, y_test = train_test_split(x,df2_y,random_state = r_state,test_size=0.20)
    lr = linear_model.LinearRegression()
    lr.fit(x_train,y_train)
    y_pred = lr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print("Maximum R2 score is achived for random state ",final_r_state," with a score of",max_r_score)
```

```python
# cross validation

score=cross_val_score(lr,x,df2_y,cv=5,scoring="r2")
abs(score).mean()
```

```python
# Cross validation explicitly

from sklearn.model_selection import KFold
folds = KFold(n_splits = 6, shuffle = True, random_state = 69)
scores = cross_val_score(lr, x_train, y_train, scoring='r2', cv=folds)
abs(scores).mean()
```

```python
# Final model

x_train,x_test,y_train,y_test=train_test_split(x,df2_y,test_size=0.2,random_state=69)

print("The number of data going to train the model is", x_train.shape,y_train.shape)
print("The number of data going to test the model is", x_test.shape,y_test.shape)
```

```python
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print("R2 score is : ",r2_score(y_test,y_pred))
print("RMSE is : ",np.sqrt(mean_squared_error(y_test,y_pred)))
```

```python
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))
    ax1 = sns.distplot(RedFunction, hist=False, color="r", label=RedName)
    ax2 = sns.distplot(BlueFunction, hist=False, color="b", label=BlueName, ax=ax1)
    plt.title(Title)
    plt.xlabel('Input features')
    plt.ylabel('Number of wins')
    plt.show()
    plt.close()
    
def PollyPlot(xtrain, xtest, y_train, y_test, lr,poly_transform):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))
    xmax=max([xtrain.values.max(), xtest.values.max()])
    xmin=min([xtrain.values.min(), xtest.values.min()])
    x=np.arange(xmin, xmax, 0.1)
    plt.plot(xtrain, y_train, 'ro', label='Training Data')
    plt.plot(xtest, y_test, 'go', label='Test Data')
    plt.plot(x, lr.predict(poly_transform.fit_transform(x.reshape(-1, 1))), label='Predicted Function')
    plt.ylim([-10000, 60000])
    plt.ylabel('Price')
    plt.legend()
```

```python
# Prediction using train Data

yh_train = lr.predict(x_train)

#Prediction using test Data

yh_test = lr.predict(x_test)
```

```python
Title = 'Distribution  Plot of  Predicted Value Using Training Data vs Training Data Distribution'
DistributionPlot(y_train, yh_train, "Actual Values (Train)", "Predicted Values (Train)", Title)
```

```python
Title='Distribution  Plot of  Predicted Value Using Test Data vs Data Distribution of Test Data'
DistributionPlot(y_test,yh_test,"Actual Values (Test)","Predicted Values (Test)",Title)
```
